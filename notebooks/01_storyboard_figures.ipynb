# Storyboard - Figures principales

Notebook pour visualiser et composer les figures principales du rapport

```python
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from PIL import Image

%matplotlib inline
sns.set_style('whitegrid')

figures_dir = Path('../reports/figures')
```

## 1. Statistiques descriptives

```python
# Charger et afficher les figures de stats descriptives
img = Image.open(figures_dir / 'class_distribution.png')
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis('off')
plt.title('Distribution des classes')
plt.show()
```

## 2. Analyse ACM

```python
# Scree plot
if (figures_dir / 'mca_scree_plot.png').exists():
    img = Image.open(figures_dir / 'mca_scree_plot.png')
    plt.figure(figsize=(10, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title('Scree plot - Inertie expliqu√©e')
    plt.show()
```

```python
# Plan factoriel
if (figures_dir / 'mca_individuals_plot.png').exists():
    img = Image.open(figures_dir / 'mca_individuals_plot.png')
    plt.figure(figsize=(10, 8))
    plt.imshow(img)
    plt.axis('off')
    plt.title('Plan factoriel - ACM')
    plt.show()
```

## 3. Clustering

```python
# Dendrogramme et clusters
if (figures_dir / 'cah_dendrogram.png').exists():
    img = Image.open(figures_dir / 'cah_dendrogram.png')
    plt.figure(figsize=(12, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title('Dendrogramme - CAH')
    plt.show()
```
